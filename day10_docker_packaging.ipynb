# Day 10 — Docker packaging

Write `requirements.txt` and `Dockerfile` for portable deployment.

````python
# === Setup ===
from google.colab import drive
drive.mount('/content/drive')

import os, re, time, uuid
import numpy as np
import pandas as pd
from tqdm import tqdm

BASE = "/content/drive/MyDrive/mimic_project"
print("BASE:", BASE)
````

````python
req_text="\n".join(["fastapi","uvicorn","python-multipart","pandas","numpy","torch","transformers","pillow","faiss-cpu",""])
open(f"{BASE}/requirements.txt","w").write(req_text)
print("Wrote:", f"{BASE}/requirements.txt")
````

````python
dockerfile=r"""FROM python:3.10-slim
WORKDIR /app
COPY requirements.txt /app/requirements.txt
RUN pip install --no-cache-dir -r /app/requirements.txt
COPY app.py /app/app.py
EXPOSE 8000
CMD [\"uvicorn\",\"app:app\",\"--host\",\"0.0.0.0\",\"--port\",\"8000\"]
"""
open(f"{BASE}/Dockerfile","w").write(dockerfile)
print("Wrote:", f"{BASE}/Dockerfile")
````

Build/run locally:

```bash
docker build -t radiology-copilot .
docker run -p 8000:8000 radiology-copilot
curl -s http://127.0.0.1:8000/health
```
