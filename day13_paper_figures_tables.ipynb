```python
# === Setup ===
from google.colab import drive
drive.mount('/content/drive')

import os, re, time, uuid
import numpy as np
import pandas as pd
from tqdm import tqdm

BASE = "/content/drive/MyDrive/mimic_project"
print("BASE:", BASE)
```

```python
import matplotlib.pyplot as plt
```

```python
e = pd.read_csv(f"{BASE}/eval_day12_rag.csv")
fig_dir = f"{BASE}/figures"
os.makedirs(fig_dir, exist_ok=True)

plt.figure(); plt.hist(e.best_score, bins=30); plt.xlabel("Best score"); plt.ylabel("Count"); plt.title("Retrieval score distribution")
plt.savefig(f"{fig_dir}/retrieval_score_hist.png", dpi=200, bbox_inches="tight"); plt.close()

plt.figure(); plt.hist(e.citation_coverage, bins=10); plt.xlabel("Citation coverage"); plt.ylabel("Count"); plt.title("Citation coverage")
plt.savefig(f"{fig_dir}/citation_coverage_hist.png", dpi=200, bbox_inches="tight"); plt.close()

rates = e.status.value_counts(normalize=True)
plt.figure(); plt.bar(rates.index.astype(str), rates.values); plt.ylabel("Fraction"); plt.title("Refusal vs Generated")
plt.savefig(f"{fig_dir}/refusal_rate_bar.png", dpi=200, bbox_inches="tight"); plt.close()

summary = pd.DataFrame([{
  "n": len(e),
  "refusal_rate": float((e.status=='refused').mean()),
  "avg_best_score": float(e.best_score.mean()),
  "avg_citation_coverage": float(e.citation_coverage.mean())
}])
summary.to_csv(f"{BASE}/paper_results_summary.csv", index=False)
print("Saved figures to:", fig_dir)
summary
```
